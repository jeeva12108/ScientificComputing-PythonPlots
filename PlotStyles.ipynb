```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10) #create array of linear apced number from 0 to 10
#print(x)
fig, ax = plt.subplots() #object creation for figure and axes

plt.style.use('fivethirtyeight') #plot style

for n in range(-20,30,10):
    ax.plot(x, np.cos(x) + np.random.randn(50) + n)

ax.set_title("'fivethirtyeight' style")

plt.show()
```
