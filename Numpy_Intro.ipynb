```python
#We can use %timit to compare a mathematical operation on a Python list using a for loop to the same mathematical 
#operation on a NumPy array. 
```

```python
lst = list(range(10000))
%timeit for i, item in enumerate(lst): lst[i] = lst[i]*2
```
