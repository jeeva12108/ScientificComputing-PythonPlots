```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.random.randn(150) #generates 150 ramdom number
#print(x1) #check the impact without seeding
y1=np.random.randn(150)
x2=np.random.randn(150)+4
y2=np.random.randn(150)+4

x=np.append(x1,x2)
y=np.append(y1,y2)

fig, ax=plt.subplots()
ax.scatter(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(1216584) #With the seed reset (every time), the same set of numbers will appear every time.
#If the random seed is not reset, different numbers appear with every invocation:


N = 50
x = np.random.rand(N)
#print(x) #check the impact of seeding
y = np.random.rand(N)
#print(y)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
