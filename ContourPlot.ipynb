```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(-4.0,4.0,0.1) #points for x axis
y = np.arange(-4, 4, 0.1) #points for y axis

X, Y = np.meshgrid(x, y)

Z = np.sin(X)*np.cos(Y)
#the depth of the plot - this can be temp / pressure on the surface for eg.

fig, ax = plt.subplots(figsize=(6,6))

mycmap1 = plt.get_cmap('gnuplot2') #check matplotlib for different color schemes fot cmap
cf = ax.contourf(X,Y,Z,cmap=mycmap1)
fig.colorbar(cf,ax=ax)

ax.set_aspect('equal')
#Where cf is the plot object created by ax.contourf(X, Y, Z)
#axis object that contains the contour plot, ax is passed to the fig.colorbar() method along with the cf plot object.

plt.show()

#Color bars are added to Matplotlib contour plots using the fig.colorbar() method. 
#Since the color bar is not part of the contour plot, the color bar needs to be applied to the figure object, 
#often called fig. A contour plot needs to be passed into the fig.colorbar() method. Therefore, 
#when you add a color bar to a figure, a plot object needs to be available. 
#A plot object is the output of the ax.contourf() method.
# complete list is available in the Matplotlib documentation 
```
