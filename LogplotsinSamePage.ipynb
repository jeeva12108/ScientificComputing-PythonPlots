```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0.01,20, 0.01)

# Create a figure with 2 rows and 2 cols of subplots
fig,((ax1,ax2), (ax3,ax4))=plt.subplots(2,2) 

#note 4 objects are created for axes with a single figure.
#the ax objects has to be grouped in a way u need in the figure window.
#the subplot function to have argument that is equivalent to the ax object declaration.

#linear plot
ax1.plot(t,np.exp(-t/50)) #note that the y axis value is directly mentioned in the argument
ax1.set_title('Linear Plot')
ax1.grid(True)

# log y axis
ax2.semilogy(t, np.exp(-t / 5.0)) #note semilogy instead of plot function
ax2.set_title('semilogy')
ax2.grid()

# log x axis
ax3.semilogx(t, np.exp(-t / 5.0)) #note semilogx instead of plot function
ax3.set_title('semilogx')
ax3.grid()

# log x and y axis
ax4.loglog(t, 20 * np.exp(-t / 5.0), basex=2) #note loglog instead of plot function
ax4.set_title('loglog base 2 on x')
ax4.grid()

fig.tight_layout()
plt.show()

```
