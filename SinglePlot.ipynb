```python
import matplotlib.pyplot as plt #import package
import numpy as np
x=np.arange(0,4*np.pi,0.1) #generate an np array start:0, stop:4pi, increament:0.1
y=np.sin(x)#create an np array which is sin prodcut of array x
#z=np.sin(x)*2 check point
plt.plot(x,y,linewidth=3.0,linestyle='dashed',color="#FF69B4",alpha=0.5,marker='o') #plot with attributes
plt.title('Sinplot')
plt.xlabel('Radian')
plt.ylabel('sin(x)')
#plt.legend(['Sin function values']) check point
plt.grid(True)
#plt.savefig('plot.pdf', dpi=300, bbox_inches='tight')
plt.show() 
#Note that plt.show() needs to be called after plt.plot() and any plot details such as plt.title()
#The plt.savefig() function needs to be called right above the plt.show() line. 
#All the features of the plot must be specified before the plot is saved as an image file. 
```
