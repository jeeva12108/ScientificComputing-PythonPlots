```python
import numpy as np
import matplotlib.pyplot as plt

Dep=["Chem","FCC","NAO","Hydro"] #each piece of pie
sizes=[114,30,40,20] #portion of that pie

fig,ax=plt.subplots() #object created
explode = (0.1, 0.1, 0.1, 0.4) # Explode out the 'Hydro' pie piece by offsetting it a greater amount
ax.pie(sizes,
       explode=explode,
       labels=Dep,
       autopct='%1.1f%%',
       shadow=True,
       startangle=90)
ax.axis('equal') # Equal aspect ratio ensures the pie chart is circular.
ax.set_title("strength")
plt.show()
```
