```python
#each type of plot will use a certain jorgan specific to that plot type.maek sure you remember that
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data - the Co.eff of thermal explansions of solid materials: lab data
aluminum = np.array([
    6.4e-5, 3.01e-5, 2.36e-5, 3.0e-5, 7.0e-5, 4.5e-5, 3.8e-5, 4.2e-5, 2.62e-5,
    3.6e-5
])
copper = np.array([
    4.5e-5, 1.97e-5, 1.6e-5, 1.97e-5, 4.0e-5, 2.4e-5, 1.9e-5, 2.41e-5, 1.85e-5,
    3.3e-5
])
steel = np.array([
    3.3e-5, 1.2e-5, 0.9e-5, 1.2e-5, 1.3e-5, 1.6e-5, 1.4e-5, 1.58e-5, 1.32e-5,
    2.1e-5
])
```

```python
alum_mean=np.mean(aluminum)
cu_mean=np.mean(copper)
st_mean=np.mean(steel)
```

```python
#creating array for plot
materials=["Aluminium", "Copper","Steel"]
x_pos=np.array(materials)
CTEs=[alum_mean, cu_mean, st_mean]
#by this the labels for the bar chart is ready
```

```python
fig, ax = plt.subplots() #object for plat- the base plate is ready

ax.bar(x_pos, CTEs, align='center', alpha=0.5,color='m') #the first argument is x positions and second is the height
ax.set_xlabel("materials")
ax.set_ylabel("COTE in (C-1)")
ax.set_title('Coefficent of Thermal Expansion (CTE) of Three Metals')
ax.set_xticks(x_pos)
ax.yaxis.grid(True) #note you can just use one axis as grid!
plt.tight_layout()
#plt.savefig('bar_plot.png')
plt.show()
```
